## Exercise 09.1 (checking data validity)

The Fibonacci series is valid only for $n \ge 0$. Add to the Fibonacci function in this notebook a check that raises an exception if $n < 0$. Try some invalid data cases to check that an exception is raised.

*Optional:* Use `pytest` to test that an exception *is* raised for some $n < 0$ cases.

```python
def f(n): 
    "Compute the nth Fibonacci number using recursion"
    # YOUR CODE HERE
    if n<0:
      raise ValueError("Input must be positive.")
    elif n <= 1:
      return n
    else:
      return(f(n-1) + f(n-2))
print(f(15))
    
      
      
```

```python
# Perform some tests    
assert f(0) == 0
assert f(1) == 1
assert f(2) == 1
assert f(3) == 2
assert f(10) == 55
assert f(15) == 610

# Check that ValueError is raised for n < 0
import pytest
with pytest.raises(ValueError):
    f(-1)
with pytest.raises(ValueError):
    f(-2)
```

```python
def f(x):
    "Evaluate polynomial function"
    return x**3 - 6*x**2 + 4*x + 12

def compute_root(f, x0, x1, tol, max_it):
    # YOUR CODE HERE
    if type(max_it) != int:
      raise ValueError("Maximum number of iterations must be an integer.")
    if max_it <0:
      raise ValueError("Maximum number of iterations must be positive.")
    error = tol + 1
    it = 0
    while error >= tol:
        # Compute midpoint
        x_mid = (x0 + x1)/2
        f_mid = f(x_mid)
        f0 = f(x0)
        # Evaluate function at left end-point and at midpoint
        error = abs(f_mid)
        if f0*f_mid < 0:
            x1 = x_mid
        else:
            x0 = x_mid
        it = it + 1
        # Guard against an infinite loop
        if it >= max_it:
            raise ValueError("Maximum number of iterations exceeded.")
            break 
    f_mid = f(x_mid)
    return x_mid, f_mid, it


```
